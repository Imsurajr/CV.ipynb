```python
#Gauss Naive, Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
data = pd.read_csv('seattle-weather.csv')
data.head(5)
```

```python
data.shape
```

```python
#Data Statiscs
data.describe()
```

```python
data.isnull().any()
```

```python
#Features of data
features = ["precipitation", "temp_max", "temp_min", "wind",]
X = data[features]
y= data.iloc[:, 5]
```

```python
X.head(5)
```

```python
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=0)
 X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#Model Building - GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
#Model Prediction
y_preds = gnb.predict(X_test)
y_preds
```

```python
#Accuracy Score
print("Training Score (accuracy): ", accuracy_score(y_train, gnb.predict(X_train)))
print("Testing Score  (accuracy): ", accuracy_score(y_test, gnb.predict(X_test)))
```

```python
#Confusion Matrix
confusion = confusion_matrix(y_test, y_preds, labels=gnb.classes_)
display = ConfusionMatrixDisplay(confusion_matrix=confusion, display_labels=gnb.classes_)
display.plot()
```

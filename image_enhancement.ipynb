```python
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def add_noise(img):
    row, col = img.shape[:2]
     # Obtain the image dimensions
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        ycord = random.randint(0, row - 1)
        xcord = random.randint(0, col - 1)
        img[ycord, xcord] = 255
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        ycord = random.randint(0, row - 1)
        xcord = random.randint(0, col - 1)
        img[ycord, xcord] = 0
    return img
```

```python

img = cv2.imread('/content/unnamed.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

noisy_img = add_noise(np.copy(img_rgb))

plt.imshow(noisy_img)
plt.show()
```

```python
# ENHANCING THE IMAGE

median = cv2.medianBlur(noisy_img, 5)
plt.imshow(median)
plt.show()
```

```python
im = cv2.imread('/content/output.jpg',0)
imginv = 255 - im

fig = plt.figure(figsize = (15,5))
ax = fig.add_subplot(121)
plt.title('Original', fontsize = 18)
plt.imshow(im,cmap='gray')
ax = fig.add_subplot(122)
plt.imshow(imginv, cmap='gray')
ax.set_title('Inverse Transform', fontsize = 18)
```

```python
'''Log Transformation'''


im = cv2.imread('/content/output.jpg',0)
log = 0.6 * (np.log(1 + np.float32(im)))
fig = plt.figure(figsize = (15,5))
ax = fig.add_subplot(121)
plt.title('Original',fontsize=18)
plt.imshow(im, cmap = 'gray')
ax = fig.add_subplot(122)
plt.imshow(log, cmap = 'gray')
ax.set_title('log Transform', fontsize = 18)
```

```python
''' Gamma Trnasformation'''

im = cv2.imread('/content/output.jpg',0)
im1 = im/255.0
im_power_law_transformation = cv2.pow(im1, 1.8)
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
plt.title('Original', fontsize =  18)
plt.imshow(im, cmap = 'gray')
ax = fig.add_subplot(122)
plt.imshow(im_power_law_transformation, cmap = 'gray')
ax.set_title('Power Law Transform',fontsize = 18)
```

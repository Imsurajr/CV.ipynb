```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('opencv.png')
imgRGB = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
img_Gray = cv2.cvtColor(imgRGB , cv2.COLOR_RGB2GRAY)
img_blur = cv2.GaussianBlur(img_Gray , (3, 3) , 0)
plt.title("Original Image")
plt.imshow(imgRGB)
```

```python
edges = cv2.Canny(image = img_blur , threshold1 = 100 , threshold2 = 200)
edges = cv2.cvtColor(edges , cv2.COLOR_BGR2RGB)
plt.imshow(edges)
```

```python
# HARRIS CORNER DETECTION

img = cv2.imread('/content/demoblockk.jpeg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray = np.float32(img_gray)
plt.imshow(img)
```

```python

corner_img = cv2.cornerHarris(img_gray, 2, 5, 0.07)
corner_img = cv2.dilate(corner_img, None)
img[corner_img > 0.01 * corner_img.max()] = [0, 0, 255]
plt.imshow(img)
```

```python
img = cv2.imread('road.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
edges = cv2.Canny(gray, 50, 150, apertureSize= 3)

lines = cv2.HoughLines(edges, 1,np.pi/180, 200)

if lines is not None:
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

plt.imshow(img)
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/opencv.png')
name = 'OpenCV Python Image'
plt.title(name)
plt.imshow(img)
```

```python
#flipping horizontally

horizontl_flip = cv2.flip(img,1)
name = 'OpenCV Pyhton Flip Image Horizontally'
plt.title(name)
plt.imshow(horizontl_flip)
```
